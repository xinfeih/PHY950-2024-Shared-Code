# Legendre
Example of first few [Legendre Polynomials](https://en.wikipedia.org/wiki/Legendre_polynomials).

---

```python
import ROOT
```

Create the canvas.

```python
canvas = ROOT.TCanvas("DistCanvas", "Legendre polynomials example", 10, 10, 750, 600)
canvas.SetGrid()
```

Load the `MathMore` shared library (automatic loading of libraries does is not yet supported for functions).

```python
ROOT.gSystem.Load("libMathMore");
```

Create the set of Legendre functions.

```python
legendrePols=[]
for i in range(5):
    lpol = ROOT.TF1("L_{%d}(x)"%i, "ROOT::Math::legendre([0],x)", -1, 1)
    lpol.SetParameters(i, 0.0);
    lpol.SetLineStyle(1);
    lpol.SetLineWidth(2);    
    lpol.SetLineColor(i+1)
    legendrePols.append(lpol)

legendrePols[0].SetMaximum(1.19)
legendrePols[0].SetMinimum(-1.19)
legendrePols[0].SetTitle("Legendre polynomials")
```

Create the legend and draw the functions.

```python
leg = ROOT.TLegend(0.75, 0.13, 0.85, 0.33)
isFirst = True
for lpol in legendrePols:
    lpol.Draw("" if isFirst else "same")
    isFirst = False
    leg.AddEntry(lpol, lpol.GetName(), "l")

leg.Draw()
```

Draw the canvas.

```python
canvas.Draw()
```
